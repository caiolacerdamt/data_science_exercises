```python
import numpy as np
import pandas as pd

np.random.seed(42)
data = {
    "idade": [25, 30, np.nan, 40, 35, np.nan, 50],
    "salario": [3000, 4000, 5000, np.nan, 6000, 7000, np.nan],
    "experiencia": [1, 5, 3, np.nan, 8, 10, np.nan]
}

df = pd.DataFrame(data)
df.head()
```

```python
df["Idade_missing"] = df["idade"].isnull().astype(int)
df["Salario_missing"] = df["salario"].isnull().astype(int)
df["Experiencia_missing"] = df["experiencia"].isnull().astype(int)

df.head()
```
